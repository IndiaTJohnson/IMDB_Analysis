#Questions and Visualization

How many movies had at least some valid financial information (values > 0 for budget OR revenue)?
Please exclude any movies with 0's for budget AND revenue from the remaining visualizations 3558 movies.

```python
#drop rows with 0 for or less for budget
df.drop(df[df.budget <= 0].index, inplace=True)
```

```python
#drop rows with 0 or less for revenue
df.drop(df[df.revenue <= 0].index, inplace=True)
```

```python
df
```

How many movies are there in each of the certification categories (G/PG/PG-13/R)?

```python
#Plot number of movies in each certification type
ax = sns.countplot(data=df, x = 'certification')
```

What is the average revenue per certification category?

```python
average_rev = df.groupby(['certification']).mean()["revenue"]#https://www.geeksforgeeks.org/pandas-groupby-multiple-values-and-plotting-results/
```

```python
average_rev
```

```python
#Plot revenue by certification type
ax = sns.barplot(data=df, x='certification', y='revenue')
```

What is the average budget per certification category?

```python
#Plot budget by certification type
ax = sns.barplot(data=df, x='certification', y='budget')
```
